```python
import cv2
import os
import time
```

```python
path1 = r'samplevideo.mp4'
cap = cv2.VideoCapture(path1)
```

### Creating dir to store images

```python
try:
    if not os.path.exists('data'):
        os.makedirs('data')

except Exception as e:
    print(f'{e} Occured')
    
```

### Frame variable and video_name

```python
total_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)

fps = int(cap.get(cv2.CAP_PROP_FPS))

desired_frames = total_frames // fps

videos = path1.split('.')
video_name = videos[0]

desired_frames

curr_frame = 0
frame_count = 0
```

### 1] To get all the frames of the video

```python
while True:
    
    ret ,frame = cap.read()
    if ret:
        name = './data/' + video_name + '_frame' +str(curr_frame) + '.jpg'
        
        cv2.imwrite(name,frame)
        curr_frame += 1
        
    else:
        print(f'Created {curr_frame} images')
        break

cap.release()
cv2.destroyAllWindows()
```

### 2] TO get as much frame as much the seconds of video

```python
path1 = r'samplevideo.mp4'
cap = cv2.VideoCapture(path1)

try:
    if not os.path.exists('datas'):
        os.makedirs('datas')

except Exception as e:
    print(f'{e} Occured')
    

fps = int(cap.get(cv2.CAP_PROP_FPS))


videos = path1.split('.')
video_name = videos[0]


curr_frame = 0
frame_count = 1

print(total_frames,fps,desired_frames)
```

```python
while True:
    
    ret ,frame = cap.read()
    
    if ret:
        if frame_count % fps == 0:
            name = './datas/' + video_name + '_frame' +str(curr_frame) + '.jpg'
            
            cv2.imwrite(name,frame)
            curr_frame += 1
            frame_count += 1
        else:
            frame_count += 1
        
    else:
        print(f'Created {curr_frame} images')
        break

cap.release()
cv2.destroyAllWindows()
```

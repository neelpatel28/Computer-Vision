# EROSION & DILATION
## Erosion - It diminishes the features of an image , and erodes the boundary of the object.
## Dilation - It increases the object area ,and used to accentuate features of the image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('saved/tree.jpg')
```

```python
kernel = np.ones((5,5),np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=1) 
img_dilation = cv2.dilate(img, kernel, iterations=1) 


cv2.imshow('Input', img) 
cv2.imshow('Erosion', img_erosion) 
cv2.imshow('Dilation', img_dilation) 
  
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(1,3,figsize=(18,18))
ax[0].imshow(img_erosion[:,:,::-1])
ax[0].set_title('Image Erosion')
ax[0].axis('off')

ax[1].imshow(img[:,:,::-1])
ax[1].set_title('Original Image')
ax[1].axis('off')

ax[2].imshow(img_dilation[:,:,::-1])
ax[2].set_title('Image Dilation')
ax[2].axis('off')
```

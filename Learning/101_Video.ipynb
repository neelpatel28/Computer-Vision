```python
cap = cv2.VideoCapture(0)

```

## 1] Open the webcam video on screen

```python
def live_video():
    while True:
        ret, frame = cap.read()  
        time.sleep(1/20) # to increase frame rate
        # gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # in-case we want black and white video , also change below 
        
        frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
        frame_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        frame_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        cv2.imshow('Video',frame)
        if cv2.waitKey(1) & 0xFF == ord('x'):
            break
    print(f'''frame_count : {frame_count}
frame_width : {frame_width}
frame_height : {frame_height}''' ,)
    cap.release()   
    cv2.destroyAllWindows()
```

```python
live_video()
```

## 3] Saving Live Video

```python
def record_video():
    cap = cv2.VideoCapture(0)
    
    if not os.path.exists('./Recordings'):
        os.makedirs('./Recordings')
    
    fourcc = cv2.VideoWriter_fourcc(*'XVID') # Error ?
    out = cv2.VideoWriter('./Recordings/liverecording.avi', fourcc, 20.0, (640, 480))
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        time.sleep(1/20) # to increase frame rate
        # gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # in-case we want black and white video , also change below 
        
        out.write(frame)  # Write the frame to the video file
        cv2.imshow('Video', frame)

        if cv2.waitKey(1) & 0xFF == ord('x'):
            break
    
    cap.release()
    out.release()   
    cv2.destroyAllWindows()
```

```python
record_video()
```

## 4] Canny vs Original Video

```python
import cv2
import numpy as np
import time
```

```python
def canny_video():
    path2 = r'samplevideo.mp4'
    cap = cv2.VideoCapture(path2)
    while True:
        ret, frame = cap.read()  
        time.sleep(1/20) # to increase frame rate
        
        gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        canny_img = cv2.Canny(gray_img,125,175)
        
        canny_img_3d = cv2.cvtColor(canny_img,cv2.COLOR_GRAY2BGR)
        
        canny_img_3d = cv2.resize(canny_img_3d, (500,400))
        frame = cv2.resize(frame, (500,400))
        # print(frame.ndim,canny_img.ndim)
        combined_frame = np.hstack((frame, canny_img_3d))  # or use np.vstack for vertical combination 
        
        # frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
        # frame_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        # frame_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        
        cv2.imshow('Original vs Canny ',combined_frame)
        if cv2.waitKey(1) & 0xFF == ord('x'):
            break
    cap.release()   
    cv2.destroyAllWindows()
    
canny_video()
```

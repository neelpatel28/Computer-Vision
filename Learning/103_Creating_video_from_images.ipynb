```python
import os
import cv2
from PIL import Image
import numpy as np
```

```python
try:
    if not os.path.exists('SavedVideo'):
        os.makedirs('SavedVideo')

except Exception as e:
    print(f'{e} Occured')
    
print(os.getcwd())
```

```python
mean_height = 0
mean_width = 0

path = './data' # path where photos are present
dir = os.listdir(path) 
total_img = len(dir) # total number of files in dir

# print(dir)

# getting image from dir and getting it's height and width
for file in dir:
    img = Image.open(os.path.join(path,file))
    width, height = img.size 
    mean_width += width 
    mean_height += height
    # img.show()

mean_width = int(mean_width/total_img)
mean_height = int(mean_height/total_img)

print(f'''mean_width : {mean_width}
mean_height : {mean_height}
      ''')

print(os.getcwd())
```

```python
def VideoGenerator():
    image_folder = './data'
    video_name = 'HDvideo.mp4'
    
    images = [img for img in os.listdir(image_folder)]
    # print(images)
    
    frame = cv2.imread(os.path.join(image_folder,images[1]))
    
    height, width, layers = frame.shape   
  
    # os.chdir('./SavedVideo')
    print(os.getcwd())
    
    fourcc = cv2.VideoWriter_fourcc(*'mp4v') # XVID -> avi
    # video = cv2.VideoWriter(video_name, 0, 1, (width, height)) 
    video = cv2.VideoWriter(video_name,fourcc, 30, (width, height)) 
  
    for image in images:  
        video.write(cv2.imread(os.path.join(image_folder, image)))  
      
    cv2.destroyAllWindows()  
    video.release()
    os.chdir('..')
    print(os.getcwd())
    
```

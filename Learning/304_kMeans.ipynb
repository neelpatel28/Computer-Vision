```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def kMeans(img,K):
    img = cv2.imread(f'saved/{img}.jpg')

    img1 = img.reshape((-1,3))
    img1 = np.float32(img1)

    # define criteria, number of clusters(K) and apply kmeans()
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    # print(cv2.TERM_CRITERIA_COUNT)
    # print(cv2.TERM_CRITERIA_EPS , cv2.TERM_CRITERIA_MAX_ITER)
    ret,label,center=cv2.kmeans(img1,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
    # Now convert back into uint8, and make original image
    center = np.uint8(center)
    res = center[label.flatten()]
    res2 = res.reshape((img.shape))
    
    res2 = cv2.cvtColor(res2,cv2.COLOR_BGR2RGB)
    return res2
```

```python
name_of_file = 'rj'
og = cv2.imread(f'saved/{name_of_file}.jpg')
og = cv2.cvtColor(og,cv2.COLOR_BGR2RGB)

k2 = kMeans(name_of_file,2)
k4 = kMeans(name_of_file,4)
k8 = kMeans(name_of_file,8)


fig,ax = plt.subplots(2,2,figsize=(18,18))

ax[0,0].imshow(og)
ax[0,0].axis('off')
ax[0,0].set_title('Original')

ax[0,1].imshow(k2)
ax[0,1].axis('off')
ax[0,1].set_title('k = 2')

ax[1,0].imshow(k4)
ax[1,0].axis('off')
ax[1,0].set_title('k = 4')

ax[1,1].imshow(k8)
ax[1,1].axis('off')
ax[1,1].set_title('k = 8')

,
```

# Feature Extraction

## It is a process of checking important features of the images

##### Features can be -> edges, corners, ridges & blobs in the image

####  
####
### 1] Edges : 
#### Edges are boundaries between different regions in an image, often corresponding to significant changes in intensity or color.
####
#### Application : Identifying the structure of objects within the image.
####
#### Techniques : Hough Detection, Sobel Operator and Canny Edge Detector
##

####  
####
### 2] Corners : 
#### Corners are points in an image where two or more edges meet. 
####
#### Application : Image matching and Object tracking.
####
#### Techniques : Harris Corner Detector and Shi-Tomasi Corner Detection
##

####  
####
### 3] Ridges : 
#### Ridges are elongated structures in an image that often represent elongated features like veins in leaves, roads, or rivers in aerial images.
####
#### Application : Scene Understanding, Object detection.
####
#### Techniques : Hessian Matrix and Scale-Space Ridge Detection
##

####  
####
### 4] Blobs : 
#### Blobs are regions in an image that are either brighter or darker than their surroundings. 
####
#### Application : Texture Analysis, Object detection.
####
#### Techniques : Difference of Gaussians (DoG),Laplacian of Gaussian (LoG) and Determinant of Hessian (DoH)
##

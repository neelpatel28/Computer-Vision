# IMAGE PYRAMIDS

#### Lvl 0 : Original Image
#### Lvl 1 : 0.5 Resolution [ 1 / 2 ] 
#### Lvl 2 : 0.25 Resolution [ 1 / 4 ] 
#### Lvl 3 : 0.125 Resolution [ 1 / 8 ] 
#### Lvl 4 : 0.0625 Resolution [ 1 / 16 ] 

Applications - Img Blending & Img Composting

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

pyrUp: This function is used to upscale an image. It increases the size of an image by interpolating the pixels. When you pyrUp an image, it doubles the image dimensions along each axis, effectively increasing the image size by four times.

pyrDown: Conversely, 'pyrDown' is used to downscale an image. It reduces the size of an image by a factor of two along each axis. This function applies a Gaussian blur and then subsamples the image, resulting in a smoother, lower-resolution version of the original image.

pyrDown -> Scales down the Image size-wise and decreases the Size of the image with each iteration ( Size Decreases )

```python
img = cv2.imread('saved/virat1.jpg',-1)
img = cv2.resize(img,(1000,800))
lr1 = cv2.pyrDown(img)
lr2 = cv2.pyrDown(lr1)
lr3 = cv2.pyrDown(lr2)

cv2.imshow('Original Image',img)
cv2.imshow('pyrDown I Image',lr1)
cv2.imshow('pyrDown II Image',lr2)
cv2.imshow('pyrDown III Image',lr3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Below the method is same only for loop is used to avoid re-writing and matplotlib is used to view it in the notebook.

```python
img = cv2.imread('saved/virat1.jpg',-1)
img = cv2.resize(img,(1000,800))

n = int(input("Enter number of iterations required : "))

fig,ax = plt.subplots(1,n,figsize=(20,8))

for i in range(n):
    layer = cv2.pyrDown(img)
    
    layer_mp = cv2.cvtColor(layer,cv2.COLOR_BGR2RGB)
    
    ax[i].imshow(layer_mp)
    ax[i].set_title(f'pyrDown {i+1} : {layer_mp.shape}')
    # ax[i].axis('off')
    
    cv2.imshow(f'pyrDown {i+1}',layer)
    img = layer
    cv2.waitKey(0)

plt.show()
cv2.destroyAllWindows()
```

pyrUp - Scales down the quality of the image pixel wise but increases the size of image ( Size Increases Quality Decreases )

```python
img = cv2.imread('saved/virat1.jpg',-1)
img = cv2.resize(img,(100,80))

n = int(input("Enter number of iterations required : "))

fig,ax = plt.subplots(1,n,figsize=(20,8))

for i in range(n):
    layer = cv2.pyrUp(img)
    
    layer_mp = cv2.cvtColor(layer,cv2.COLOR_BGR2RGB)
    
    ax[i].imshow(layer_mp)
    ax[i].set_title(f'pyrUp {i+1} : {layer_mp.shape}')
    # ax[i].axis('off')
    
    cv2.imshow(f'pyrUp {i+1}',layer)
    img = layer
    cv2.waitKey(0)

plt.show()
cv2.destroyAllWindows()
```

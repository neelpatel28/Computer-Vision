# SIFT (Scale-Invariant Feature Transform) algorithm 

### > It is a computer vision technique used for feature detection and description. 
### > It detects distinctive key points or features in an image that are robust to changes in scale, rotation, and affine transformations.

```python

import numpy as np
import cv2
```

```python
sift = cv2.SIFT_create()

bf = cv2.BFMatcher(cv2.NORM_L2,crossCheck=True)

img1 = cv2.imread('saved\\india1.jpg')
img2 = cv2.imread('saved\\india2.jpg')

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
kp1,ds1 = sift.detectAndCompute(img1,None)
kp2,ds2 = sift.detectAndCompute(img2,None)

matches = bf.match(ds1,ds2)


matches = sorted(matches,key = lambda x:x.distance)

img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[300:600],img2,flags=None)

cv2.imshow('Matching',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

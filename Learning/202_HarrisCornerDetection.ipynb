```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# HARRIS CORNER DETECTION 

```python
image = cv2.imread('saved/virat1.jpg')
image = cv2.resize(image,(800,500)) 

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
gray_image = np.float32(gray_image) 

dst = cv2.cornerHarris(gray_image, blockSize=2, ksize=3, k=0.04) 

dst = cv2.dilate(dst, None) 
image[dst > 0.01 * dst.max()] = [0, 255, 0] 

cimage = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig,ax = plt.subplots(figsize=(10,8))
ax.imshow(cimage)
ax.set_title('Detected Image')
ax.axis('off')
plt.show()


cv2.imshow('Harris Corner Detection', image) 
if cv2.waitKey(0) & 0xff == 27: 
	cv2.destroyAllWindows()
```

# Shi-Tomasi Corner Detection

```python
image = cv2.imread('saved/virat1.jpg')
image = cv2.resize(image,(800,500)) 

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

corners = cv2.goodFeaturesToTrack(gray_image, maxCorners=50, qualityLevel=0.02, minDistance=20) 
corners = np.float32(corners) 

for item in corners: 
	x, y = item[0] 
	x = int(x) 
	y = int(y) 
	cv2.circle(image, (x, y), 6, (0, 255, 0), -1) 

cimage = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig,ax = plt.subplots(figsize=(10,8))
ax.imshow(cimage)
ax.set_title('Detected Image')
ax.axis('off')
plt.show()

cv2.imshow('Shi-Tomasi', image) 
if cv2.waitKey(0) & 0xff == 27: 
	cv2.destroyAllWindows()
```

# SIFT (Scale-Invariant Feature Transform)

```python
image = cv2.imread('saved/virat1.jpg')
image = cv2.resize(image,(800,500)) 

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

sift = cv2.SIFT_create() 
kp, des = sift.detectAndCompute(gray_image, None) 


kp_image = cv2.drawKeypoints(image, kp, None, color=( 
	0, 255, 0), flags=None) 

cimage = cv2.cvtColor(kp_image,cv2.COLOR_BGR2RGB)
fig,ax = plt.subplots(figsize=(10,8))
ax.imshow(cimage)
ax.set_title('Detected Image')
ax.axis('off')
plt.show()

cv2.imshow('SIFT', kp_image) 
if cv2.waitKey(0) & 0xff == 27: 
	cv2.destroyAllWindows()
```

# FAST algorithm

```python
image = cv2.imread('saved/virat1.jpg')
image = cv2.resize(image,(800,500)) 

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 


fast = cv2.FastFeatureDetector_create() 
fast.setNonmaxSuppression(False) 

kp = fast.detect(gray_image, None) 
kp_image = cv2.drawKeypoints(image, kp, None, color=(0, 255, 0)) 

cimage = cv2.cvtColor(kp_image,cv2.COLOR_BGR2RGB)
fig,ax = plt.subplots(figsize=(10,8))
ax.imshow(cimage)
ax.set_title('Detected Image')
ax.axis('off')
plt.show()

cv2.imshow('FAST', kp_image) 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

# ORB

```python
image = cv2.imread('saved/virat1.jpg')
image = cv2.resize(image,(800,500)) 

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

orb = cv2.ORB_create(nfeatures=2000) 
kp, des = orb.detectAndCompute(gray_image, None) 

kp_image = cv2.drawKeypoints(image, kp, None, color=(0, 255, 0), flags=0) 

cimage = cv2.cvtColor(kp_image,cv2.COLOR_BGR2RGB)
fig,ax = plt.subplots(figsize=(10,8))
ax.imshow(cimage)
ax.set_title('Detected Image')
ax.axis('off')
plt.show()

cv2.imshow('ORB', kp_image) 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors
from matplotlib import cm
```

```python
img = cv2.imread('saved/apple.jpg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

hist = cv2.calcHist([hsv],[0,1],None,[180,256],[0,180,0,256])

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(25,25))
plt.subplot(131)
plt.imshow(img_rgb)
plt.title('Original')
plt.axis('off')

plt.subplot(132)
plt.imshow(hist)
plt.title('Hist')
plt.ylabel('hue')
plt.xlabel('saturation')



lowerbound = np.array([106,128,0])
upperbound = np.array([112,200,255])

mask = cv2.inRange(hsv,lowerbound,upperbound)

plt.subplot(133)
plt.imshow(mask,cmap=cm.gist_rainbow)
plt.title('Mask')


plt.show()
```

```python
colors = ['r', 'g', 'b']

x = img[:,:,1].flatten()
y = x + img[:,:,2].flatten()

fig, ax = plt.subplots()
ax.hist2d(x, y, cmap="BuPu", bins=[50, 10], norm=mcolors.LogNorm())
plt.show()
```

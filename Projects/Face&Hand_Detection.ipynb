```python
import cv2
import time
import mediapipe as mp
```

```python
# Initializing Holistic Model from Mediapipe
mp_holistic = mp.solutions.holistic
holistic_model = mp_holistic.Holistic(
	min_detection_confidence=0.5,
	min_tracking_confidence=0.5
)

# Drawing utils
mp_drawing = mp.solutions.drawing_utils
```

```python
path = 0
capture = cv2.VideoCapture(path)

current = 0
earlier = 0

while capture.isOpened():

	ret, frame = capture.read()
	frame = cv2.resize(frame, (800, 600))
	image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

	# Using model
	image.flags.writeable = False
	results = holistic_model.process(image)
	image.flags.writeable = True

	# RGB2BGR for Recognition
	image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

	# Drawing the Facial Landmarks
	mp_drawing.draw_landmarks(
	image,
	results.face_landmarks,
	mp_holistic.FACEMESH_CONTOURS,
	mp_drawing.DrawingSpec(
		color=(0,255,0),
		thickness=1,
		circle_radius=1
	),
	mp_drawing.DrawingSpec(   
		color=(0,255,255),
		thickness=1,
		circle_radius=1
	)
	)

	# Drawing Right & Left hand Land Marks
	mp_drawing.draw_landmarks(
	image, 
	results.right_hand_landmarks, 
	mp_holistic.HAND_CONNECTIONS
	)

	mp_drawing.draw_landmarks(
	image, 
	results.left_hand_landmarks, 
	mp_holistic.HAND_CONNECTIONS
	)
	
	# Correct the image ( flip horizontally )
	image = cv2.flip(image, 1)
	
 
	current = time.time()
	fps = 1 / (current-earlier)
	earlier = current
	
     
	# Displaying FPS on the image
	cv2.putText(image, str(int(fps))+" FPS", (10, 70), cv2.FONT_HERSHEY_TRIPLEX, 1, (0,0,0), 2)

	cv2.imshow("Facial and Hand Landmarks", image)


	if cv2.waitKey(1) & 0xFF == ord('x'):
		break

capture.release()
cv2.destroyAllWindows()
```
